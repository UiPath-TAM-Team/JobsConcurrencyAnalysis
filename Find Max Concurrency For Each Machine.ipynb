```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
jobscsv = pd.read_csv("jobs.csv")
display(jobscsv)
```

```python
jobscsv.StartTime = pd.to_datetime(jobscsv.StartTime)
jobscsv.EndTime = pd.to_datetime(jobscsv.EndTime)
jobscsv['duration']=jobscsv.EndTime-jobscsv.StartTime
jobscsv
```

```python
avgDuration = jobscsv['duration'].mean()
firstStart = jobscsv['StartTime'].min()
lastEnd = jobscsv['EndTime'].max()
```

```python
print(str(avgDuration))
```

```python
jobscsv=jobscsv.sort_values(by='StartTime', ascending=True)
jobscsv
```

```python
increments = []
concurrencyDict = {}
incrementCounter = firstStart
concurrency = 0

while(incrementCounter < lastEnd):
    increments.append(incrementCounter)
    incrementCounter = incrementCounter + avgDuration/2
print(len(increments))   
lastIncrement = increments[0]
```

```python
uniqueMachines = (jobscsv['HostMachineName'].unique())
print(uniqueMachines)
uniqueMachineTuples = []
tupleCounter = 0
for uniqueMachine in uniqueMachines:
    BotMachine_df = jobscsv.loc[jobscsv['HostMachineName'] == uniqueMachine]
    machineTuple = (uniqueMachine, len(BotMachine_df))
    uniqueMachineTuples.append(machineTuple)
    print(str(len(BotMachine_df)))

uniqueMachineTuples.sort(key=lambda y: y[1])
print(uniqueMachineTuples)
```

```python
for machineTuple in uniqueMachineTuples:
    Machine_df = jobscsv.loc[jobscsv['HostMachineName'] == machineTuple[0]]
    print("Finding max concurrency for Machine " + machineTuple[0])
    for increment in increments:      
        for index, row in Machine_df.iterrows():
            if(row['StartTime'] < increment):
                if(lastIncrement < row['EndTime'] < increment):
                    concurrency = concurrency + 1
            else:
                concurrencyDict[increment] = concurrency
                concurrency = 0
                break
        lastIncrement = increment
    Time_Concurrency_df = pd.DataFrame(list(concurrencyDict.items()), columns=['TimeIncrement', 'BotConcurrency'])
    MaxConcur = Time_Concurrency_df['BotConcurrency'].max()
    print('The Max Concurrency for Machine' + machineTuple[0] + ' is ' + str(MaxConcur))
```
