```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
jobscsv = pd.read_csv("jobs.csv")
display(jobscsv)
```

```python
jobscsv.StartTime = pd.to_datetime(jobscsv.StartTime)
jobscsv.EndTime = pd.to_datetime(jobscsv.EndTime)
jobscsv['duration']=jobscsv.EndTime-jobscsv.StartTime
jobscsv
```

```python
avgDuration = jobscsv['duration'].mean()
firstStart = jobscsv['StartTime'].min()
lastEnd = jobscsv['EndTime'].max()
```

```python
jobscsv=jobscsv.sort_values(by='StartTime', ascending=True)
jobscsv
```

```python
increments = []
concurrencyDict = {}
incrementCounter = firstStart
concurrency = 0

while(incrementCounter < lastEnd):
    increments.append(incrementCounter)
    incrementCounter = incrementCounter + avgDuration
print(len(increments))   
lastIncrement = increments[0]
```

```python
uniqueMachines = (jobscsv['HostMachineName'].unique())
#print(uniqueMachines)
#for uniqueMachine in uniqueMachines:
#    print("Creating Jobs Gantt Chart for " + uniqueMachine)
#    MachineD001_df = jobscsv.loc[jobscsv['HostMachineName'] == uniqueMachine]
DEFRARPAP023_df = jobscsv.loc[jobscsv['HostMachineName'] == 'DEFRARPAP023']
for increment in increments:      
    for index, row in DEFRARPAP023_df.iterrows():
        if(row['StartTime'] < increment):
            if(lastIncrement < row['EndTime'] < increment):
                concurrency = concurrency + 1
                print("Concurrency is " + str(concurrency) )
        else:
            concurrencyDict[increment] = concurrency
            concurrency = 0
            break
    lastIncrement = increment
df = pd.DataFrame(list(concurrencyDict.items()), columns=['TimeIncrement', 'BotConcurrency'])
df['TimeIncrement'] = pd.to_datetime(df['TimeIncrement'])

x = df['TimeIncrement']
y = df['BotConcurrency']
plt.figure(figsize=(80, 80))
plt.title('Bot Concurrency Vs Avg Duration Time Increment') #title
plt.xlabel('Time Increment') #x label
plt.ylabel('Bot Concurrency') #y label
plt.scatter(x, y, s=20)
plt.savefig('BotConcurrencyScatterPlots/pic.png')


                
            
    
    
    
```
