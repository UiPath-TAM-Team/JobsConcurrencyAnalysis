```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
jobscsv = pd.read_csv("jobs.csv")
```

```python
display(jobscsv)
```

```python
jobscsv.StartTime = pd.to_datetime(jobscsv.StartTime)
```

```python
jobscsv.EndTime = pd.to_datetime(jobscsv.EndTime)
```

```python
jobscsv['duration']=jobscsv.EndTime-jobscsv.StartTime
jobscsv
```

```python
uniqueMachines = (jobscsv['HostMachineName'].unique())
print(uniqueMachines)
for uniqueMachine in uniqueMachines:
    print("Creating Jobs Gantt Chart for " + uniqueMachine)
    MachineD001_df = jobscsv.loc[jobscsv['HostMachineName'] == uniqueMachine]
    MachineD001_df=MachineD001_df.sort_values(by='StartTime', ascending=True)
    machineSpecTitle = "Gantt chart for jobs ran on " + uniqueMachine 
    fig = px.timeline(MachineD001_df, x_start="StartTime", x_end="EndTime", y="Key", title=machineSpecTitle,width=8000, height=4000)
    fig.update_yaxes(autorange="reversed") # otherwise tasks are listed from the bottom up
    fig.update_layout(
        font=dict(
            family="Courier New, monospace",
            size=30,
            color="Red"
        )
    )
    fig.update_traces(marker_color='red')
    fig.write_image(uniqueMachine + '.png')
```
